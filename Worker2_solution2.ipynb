```python
def get_name_set(path='.\data\\train\\0000000062_0000000000'):
    # Получить множество имен 
    name_list = np.array(os.listdir(path))
    for i in range(len(name_list)):
        name_list[i] = int(name_list[i][-14:-4])
    return set(name_list)

def check_frame(csv_path='.\data\\train_annot\\0000000062_0000000000.csv'):
    photo_set = get_name_set(path='.\data\\train\\0000000062_0000000000')

    names_ = ["frame_number", "object_id", "x", "y", "w", "h", "class", "species", "occlusion", "noise"]
    csv = pd.read_csv(csv_path, names=names_, header=None)
    exist_photo_set = set(csv.frame_number)
    dn_exist = photo_set.difference(exist_photo_set) 
    return exist_photo_set, dn_exist
```

```python
def open_photo(name_csv, id_csv_obj):
    name_csv = name_csv[:-4]
    name_photo = name_csv + '_' + str(id_csv_obj).zfill(10) + '.jpg'
    root_photo = os.path.join(r'.\data\train', name_csv, name_photo)
    img = Image.open(root_photo)
    return np.array(img)

def save_cut_pic(name_csv, vec_feat, arr):
    name = str(vec_feat[0]) + '_' + str(vec_feat[6]) + '_' + str(vec_feat[7]) + '_' + str(vec_feat[8]) + '_' + str(vec_feat[9]) + '.jpg' 
    final_path = os.path.join('.\\data\\train\\image', name_csv[:-4], name)
    plt.imsave(final_path, arr)
    return
```

```python
root_csv_path = '.\data\\train_annot'
for csv_file in os.listdir(root_csv_path)[:1]:
    path_csv = os.path.join(root_csv_path, csv_file)
    
    names_ = ["frame_number", "object_id", "x", "y", "w", "h", "class", "species", "occlusion", "noise"]
    data = pd.read_csv(path_csv, names=names_, header=None)
    #id_photo_csv = data.frame_number
    for row in data.index[:1]:
        # Достаем id фрейма из csv
        id_ = data.iloc[row,0]
        # Открываем фотку
        im_arr = open_photo(name_csv=csv_file, id_csv_obj=id_)
        
        # Достаем координаты
        x = data.iloc[row,2]
        y = data.iloc[row,3]
        w = data.iloc[row,4]
        h = data.iloc[row,5]
        
        # Мутиим фрейм
        frame = [(x, y), (x+w, y+h)]
        
        # Вырезаем
        cut_pic = im_arr[frame[0][1]:frame[1][1], frame[0][0]:frame[1][0]]
        
        # Выводим кусочеГ
        #
        
        # Сохраняем пикчу
        save_cut_pic(name_csv=csv_file, vec_feat=data.iloc[row], arr=cut_pic)
```
