```python
def preproc_stat(folder_name='\\train'):
    folfer_path = '.\data' + folder_name
    folder_names = os.listdir(folfer_path)[:-1] 
    
    buffer = np.array([[0, 0, 0]])
    for folder in tqdm(folder_names):
        # Путь к папке с фотками
        path_files = os.path.join(folfer_path, folder)
        file_names = os.listdir(path_files)
        for photo in file_names:
            # Собираем путь к файлу
            path = os.path.join(folfer_path, folder, photo) 
            
            # Открываем матрицу фото
            arr = np.array(Image.open(path))
            
            # Рассчитываем статитику
            mean_pic = np.mean(arr)
            var_pic = np.var(arr, ddof=1)
            
            # СОздаем вектор  среднего и дисперсии
            d_vec = np.array([mean_pic, var_pic, path])
            # Добавляем в глобальный буфер
            buffer = np.vstack((buffer, d_vec))
    # удалим нулевую строку buffer
    buffer = np.delete(buffer, 0, axis = 0)
    # Создаем датафрейм
    df = pd.DataFrame({
                        'mean' : buffer[:,0],
                        'variance' : buffer[:,1],
                        'path' : buffer[:,2]
                        })
    return df
```

```python
df = preproc_stat()
```
